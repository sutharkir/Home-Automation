# text annotation


```python
#draw plot as figure

fig,ax=plt.subplots()
x=np.linspace(0,30,100)
y=np.cos(x)

ax.plot(x,y)
ax.axis('equal')
ax.annotate('high',xy=(2*np.pi,1),xytext=(10,4),arrowprops=dict(facecolor='r'))
ax.annotate('low',xy=(7*np.pi,-1),xytext=(10,-4),arrowprops=dict(facecolor='r'))
plt.show()


```

# 3D plot

```python
#import mplpt3d toolkit,nump,matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
#3d projection
ax=plt.axes(projection='3d')
z=np.linspace(0,2,100)
y=z*np.sin(25*z)
x=z*np.sin(25*z)
ax.plot3D(x,y,z,'blue')
ax.set_title("3D graph")
plt.show()
```

# pie chart

```python
#creating data set
country=["india","pak","china","uk","afg"]
population=[130,80,140,45,30]
explode=(0.1,0.1,0,0,0)
plt.pie(population,labels=country,explode=explode)
fig=plt.figure(figsize=(25,15))
plt.show()
```

```python
#creating data set
country=["india","pak","china","uk","afg"]
population=[130,80,140,45,30]
explode=(0.1,0,0,0,0)
plt.pie(population,labels=country,explode=explode,shadow=True,autopct="%1.1f%%")
fig=plt.figure(figsize=(25,15))
plt.show()
```
